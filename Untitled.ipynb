## Univariate Analysis 

```python
numerical_columns = df.select_dtypes(include=["int64", "float64"]).columns
categorical_columns = df.select_dtypes(include=["object"]).columns
```

```python
numerical_columns
```

```python
type(numerical_columns)
```

```python
numerical_columns = list(numerical_columns)
```

```python
type(numerical_columns)
```

```python
l1 = ["ID", "CollegeID", "CollegeCityID"]
```

```python
numerical_columns = [x for x in numerical_columns if x not in l1]
```

```python
print(numerical_columns)
```

```python
categorical_columns
```

```python
# Non-visualization techniques on numercial columns
# here we are ignoring the column ID basic it was a index column , so no need of analysis on that column

for col in numerical_columns:
    if col == 'ID':
        continue
    print("-"*50)
    print("-"*5 , col + "-"*5)
    print("Min value is : ", df[col].min())
    print("Max value is : ", df[col].max())
    print("Mean value is : ", df[col].mean())
    print("Median value is : ", df[col].median())
    print("Mode value is : ", df[col].mode())
    print("Varience value is : ", df[col].std())
    print("Skewness is : ", df[col].skew())
    print("Kurtosis is : ", df[col].kurt())
    
```

```python
# Non-visualization techniques on categorical columns

for col in categorical_columns:
    print("-"*50)
    print("-"*5 , col + "-"*5)
    print(df[col].value_counts())
    print(df[col].unique())
    unique_count = df[col].nunique()
    print(f"Number of unique categories: {unique_count}")
    mode = df[col].mode()
    print(f"The most frequent category is: {mode[0]}")
    print(df[col].value_counts(normalize=True))

```

```python
# Visualization techniques on Numerical Columns

for col in numerical_columns:
    
    plt.figure(figsize=(8, 5))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f"Histogram for {col}")
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
for col in numerical_columns:
    
    plt.figure(figsize=(8, 5))
    sns.boxplot(y=df[col])
    plt.title(f"Boxplot for {col}")
    plt.show()
```

## Bivariate Analysis

### 1) Numerical vs Numerical

```python
corr_pearson = df[['Salary', '10percentage', '12percentage', 'collegeGPA']].corr(method='pearson')
corr_spearman = df[['Salary', '10percentage', '12percentage', 'collegeGPA']].corr(method='spearman')
print("Pearson Correlation:\n", corr_pearson)
print("Spearman Correlation:\n", corr_spearman)
```

```python
sns.scatterplot(x=df['collegeGPA'], y=df['Salary'])
plt.title("Scatter plot of College GPA vs Salary")
plt.show()
```

```python
sns.heatmap(df[['Salary', '10percentage', '12percentage', 'collegeGPA']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

### 2) Numerical vs Categorical 

```python
df.groupby('Degree')['Salary'].mean()
```

```python
from scipy import stats

stats.f_oneway(df[df['Degree'] == 'Engineering']['Salary'], df[df['Degree'] == 'Science']['Salary'])
```

```python
sns.boxplot(x='Degree', y='Salary', data=df)
plt.title("Boxplot of Salary by Degree")
plt.show()
```

```python
sns.barplot(x='Degree', y='Salary', data=df, estimator=np.mean)
plt.title("Mean Salary by Degree")
plt.show()
```

### 3) Categorical vs Categorical 

```python
pd.crosstab(df['Degree'], df['JobCity'])
```

```python
from scipy.stats import chi2_contingency

crosstab = pd.crosstab(df['Degree'], df['CollegeTier'])
chi2_stat, p_val, dof, ex = chi2_contingency(crosstab)
print(f"Chi-square statistic: {chi2_stat}, p-value: {p_val}")
```

```python
crosstab = pd.crosstab(df['Degree'], df['JobCity'])
sns.heatmap(crosstab, annot=True, cmap="YlGnBu")
plt.title("Heatmap of Degree vs JobCity")
plt.show()
```

```python
crosstab.plot(kind='bar', stacked=True)
plt.title("Stacked Bar Plot for Degree vs JobCity")
plt.show()
```
